```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

def operations(df):
    print('-'*30)
    print('First 5 records ...')
    print(df.head())
    
    print('-'*30)
    print('Last 5 records ...')
    print(df.tail())

    print('-'*30)
    print('Information ...')
    print(df.info())    

    print('-'*30)
    print('Describe ...')
    print(df.describe())

    print('-'*30)
    print('Null values ...')
    print(df.head())

    print('-'*30)
    for col in list(df.columns):
        print(f"non-null count of {df[col]} : {df[col].count()}")

    df.corr(numeric_only=True)
    print('-'*30)
    
    df.cov(numeric_only=True)
    print('-'*30)
    print('-'*30)
    print('-'*30)
    
files = ['claimants.csv','crime_data.csv','delivery_time.csv','Engine.csv','framingham.csv','glass.csv','NewspaperData.csv','SAT-GPA.csv','ToyotaCorolla.csv','TvMarketing.csv','Universities.csv','wine.csv']
for file in files:
    df = pd.read_csv(file)
    operations(df)
```

```python
print("----------------Toyota Corolla-------------------")
toyota = pd.read_csv('ToyotaCorolla.csv')

fuel_price = toyota.groupby('Fuel_Type')['Price'].count().reset_index()
plt.bar(fuel_price['Fuel_Type'],fuel_price['Price'])
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.title('Fuel Type vs Count')
plt.show()

plt.scatter(toyota['Mfg_Year'],toyota['Price'])
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Year vs price')
plt.show()
print()

plt.hist(toyota['KM'],edgecolor='black')
plt.xlabel('KM travelled')
plt.ylabel('Cars Count')
plt.title('KM distribution')
plt.show()
print()

Mfg_Year_Price = toyota.groupby('Mfg_Year')['Price'].mean().reset_index()
plt.plot(Mfg_Year_Price['Mfg_Year'],Mfg_Year_Price['Price'])
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.title('Year vs Average price')
plt.show()
print()

year_cars = toyota.groupby('Mfg_Year')['Id'].count().reset_index()
plt.pie(year_cars['Id'] ,labels = year_cars['Mfg_Year'])
plt.title('Year-wise distribution')
plt.show()

toyota_sub = toyota[['Price','Age_08_04','KM','Fuel_Type']]
sns.pairplot(toyota_sub,hue='Fuel_Type')
plt.show()
```

### 2. What is Supervised and Unsupervised machine learning?

**Ans.** 

**Supervised Learning :**
    A supervised learning is a type of machine learning in which, a model is built by passing dependent and independent variables. (input and output variables). We use an algorithm to learn the mapping function from the input to output.
    The goal is to approximate the mapping function so well that when you have new input data (x) that you can predict the output variables (Y) for that data.It is called supervised learning because the process of an algorithm learning from the training dataset can be thought of as a teacher supervising the learning process.
    
**Unsupervised Learning :**
    Unsupervised learning is where you only have input data (X) and no corresponding output variables. The goal for unsupervised learning is to model the underlying structure or distribution in the data in order to learn more about the data.These are called unsupervised learning because unlike supervised learning above there is no correct answers and there is no teacher supervision.Algorithms are left to their own devises to discover and present the interesting structure in the data.
